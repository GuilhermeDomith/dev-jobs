```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

import pandas as pd
```

```python
dftags = pd.read_csv('data/jobs_tags.csv')
dftags = dftags.drop(['id'],axis=1)
dftags.shape
```

```python
dftags.head()
```

```python
vectorizer = None
tags_corr = None
tfidf_tags = None

def get_vectorizer():
    return TfidfVectorizer(
                analyzer='char',
                use_idf = False
            )

def aplicar_tfidf(palavras = []):
    global vectorizer

    if not vectorizer:
        vectorizer = get_vectorizer()
        vectorizer.fit(palavras)
        
    matrix_tfidf = vectorizer.transform(palavras)
    
    return pd.DataFrame(
        matrix_tfidf.todense(),
        columns=vectorizer.get_feature_names()
    )

def obter_tag_similar(df_tags, df_tfidf_tags, df_tfidf_analizar):
    df_tfidf = df_tfidf_tags.append(df_tfidf_analizar, ignore_index=True)
    sim_matriz = cosine_similarity(df_tfidf)
    
    sim_serie = pd.Series(sim_matriz[-1]).sort_values(ascending=False)
    index_tag = df_tags.loc[sim_serie.index].iloc[1].name
    
    return df_tags.columns[index_tag]


def tags_correlacionadas(tag):
    return tags_corr[ tag ].sort_values(ascending=False)[:20]

def _inicializa():
    tags_corr = dftags.fillna(0).corr()
    tfidf_tags = aplicar_tfidf(dftags.columns)
    tfidf_analizar = aplicar_tfidf(['aj'])
    
```

```python
obter_tag_similar(dftags, tfidf_tags, tfidf_analizar)
```
